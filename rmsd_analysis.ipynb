```python
structures = [protein_noE_277_A,\
protein_noE_277_B,\
protein_noE_277_s,\
protein_noE_289_A,\
protein_noE_289_B,\
protein_E1_289_A,\
protein_E1_289_B,\
protein_E2_289_A,\
protein_E2_289_B]

names = ['277_noE_A',\
'277_noE_B',\
'277_noE_s',\
'289_noE_A',\
'289_noE_B',\
'289_E1_A',\
'289_E1_B',\
'289_E2_A',\
'289_E2_B']

numstr = len(names)
```

```python
plt.style.use('seaborn-poster')
```

```python
import MDAnalysis.analysis.rms


RMSDs = np.empty((numstr, numstr))
for i, st1 in enumerate(structures):
    for j, st2 in enumerate(structures):
        R = MDAnalysis.analysis.rms.RMSD(st1, st2, select='backbone or name CB')
        R.run()
        rmsd = R.rmsd[0,2]
        RMSDs[i,j] = rmsd
```

```python
plt.imshow(RMSDs)
plt.colorbar()

for i in range(numstr):
    for j in range(numstr):
        plt.text(j, i, round(RMSDs[i, j],2),
                       ha="center", va="center", color="w", size=15)

# We want to show all ticks...
plt.xticks(np.arange(numstr), labels = names, rotation=65, ha="right",
         rotation_mode="anchor")
plt.yticks(np.arange(numstr), labels = names)

plt.title("RMSDs of backbone + CB atoms in A")
plt.tight_layout()
plt.show()
        

plt.show()
```

```python
import MDAnalysis.analysis.rms


RMSDs = np.empty((numstr, numstr))
for i, st1 in enumerate(structures):
    for j, st2 in enumerate(structures):
        R = MDAnalysis.analysis.rms.RMSD(st1, st2, weights='mass', select=selection)
        R.run()
        rmsd = R.rmsd[0,2]
        RMSDs[i,j] = rmsd
```
